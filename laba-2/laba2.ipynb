```python
import numpy as np # ЗАДАНИЕ №1
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from ipywidgets import *
import math
```

```python
# Пользуясь непосредственным вводом значений, задайте матрицу A,
# размером 3 на 3, и вектор свободных коэффициентов b.

def create_A(N):
    matrix = list()
    for i in range(N):
        for j in range(N):
            if j == 0:
                matrix.append([])
            matrix[i].append(float(input(f"Введите элемент {i + 1} - {j + 1} матрицы A: ")))
    return matrix



def create_b(N):
    vector = list()
    for i in range(N):
        vector.append(float(input(f"Введите элемент {i + 1} вектора b: ")))
    return vector
```

```python
# Вызвав соответствующую функцию, получите решение системы 
# линейных уравнений Ax = b.
def solve(A, b):
    return np.linalg.solve(A, b)
# Убедитесь в правильности решения путём подстановки.
def check_solution(A, x, b):
    return np.allclose(A@x, b)
# Проверка
def calc_error(x, x0):
    return (x - x0).tolist()
```

Система: 
$$2x_1 + x_2 + x_3 = 2$$
$$x_1 - x_2 = -2$$
$$3x_1 - x_2 + 2x_3 = 2$$

```python
# Задание таблицы
A = np.array(create_A(3)) 
b = np.array(create_b(3))
```

```python
# Вывод
x0 = solve(A, b)
print('x0:', x0)
print('Solution is Correct:', check_solution(A, x0, b))
```

```python
# Постройте спираль, пользуясь параметрическим заданием функции в
# декартовых координатах.

def f():
    r = 1
    tht = np.linspace(1, 15* np.pi, 1000)
    x = r * np.cos(tht)
    y = r * np.sin(tht)
    z = np.log(tht)
    return (x, y, z)

X, Y, Z = f()

fig = plt.figure()
ax  = fig.add_subplot(projection='3d')
ax.plot(X, Y, Z)

plt.show()
```
